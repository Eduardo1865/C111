```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfPaises = pd.read_csv('paises.csv', delimiter=';')

print(dfPaises.columns)
```

Exercicio 1 - Por meio do dataset paises.csv, trace dois gráficos de linhas em um mesmo plano cartesiano, um mostrando a taxa de mortalidade (Deathrate) e outro a taxa de natalidade (Birthrate) dos países da América do Norte.


```python
dfPaises['Region'] = dfPaises['Region'].str.replace(" ", "").str.upper()
north_america = dfPaises[dfPaises['Region'] == 'NORTHERNAMERICA']


paises = north_america['Country'] # eixo X com nomes dos países


death_rates = north_america['Deathrate'] # eixo y com as taxas
birth_rates = north_america['Birthrate']


plt.figure(figsize=(14, 6)) # criar o gráfico


plt.plot(paises, death_rates, 'b--', label='Taxa de Mortalidade') # traçar as linhas com os nomes dos países no eixo X
plt.plot(paises, birth_rates, 'r--', label='Taxa de Natalidade')


plt.ylabel('Taxa (%)') # customização
plt.title('Taxa de Mortalidade e Natalidade dos Países da América do Norte')
plt.legend()
plt.show()
```

Por meio do dataset space.csv, trace um gráfico em barras mostrando quantas empresas espaciais diferentes os EUA e a CHINA possuem;
Dica: não se esqueça de retirar os resultados repetidos

```python
dfSpace = pd.read_csv('space.csv', delimiter=';') 

USA = dfSpace[dfSpace['Location'].str.contains('USA')] #separando todos os locations que tem USA

C = dfSpace[dfSpace['Location'].str.contains('China')]

Companias_unicasUSA = USA["Company Name"].nunique() #checando se as companias são unicas

Companias_unicasC = C["Company Name"].nunique()

paises = ['USA', 'China'] 
valores = [Companias_unicasUSA, Companias_unicasC] 

plt.bar(paises, valores, color=['blue', 'red'])

```

Por meio do dataset space.csv, trace um gráfico em torta ilustrando a porcentagem de missões da empresa Roscosmos que deram certo e que deram errado;

```python
Roscosmos = dfSpace[dfSpace['Company Name'].str.contains('Roscosmos')] #encontrando os roscosmos

nRocosmos = len(Roscosmos) # tamanho

Rosucesso =  len(Roscosmos[Roscosmos['Status Mission'].str.contains('Success')]) #sucesso

Rofracasso = len(Roscosmos[Roscosmos['Status Mission'].str.contains('Failure')]) #fracasso

Rosucessop = Rosucesso/nRocosmos*100 #porcentagens

Rofracassop = Rofracasso/nRocosmos*100

plt.pie(x=[Rosucessop,Rofracassop], labels=['% Sucesso','% Fracasso'], autopct='%1.1f%%') 
```
